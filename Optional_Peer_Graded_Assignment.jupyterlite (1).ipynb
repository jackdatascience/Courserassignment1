#### Read the dataset in the csv file from the URL


```python
from js import fetch
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/boston_housing.csv'
resp = await fetch(URL)
boston_url = io.BytesIO((await resp.arrayBuffer()).to_py())
```

```python
boston_df=pd.read_csv(boston_url)
```

```python
boston_df.boxplot(column=['MEDV'], grid = False)

pyplot.title('Median value of owner-occupied homes')
pyplot.show()
```

```python
sns.barplot(y=boston_df['CHAS'])
pyplot.title('Barplot for Charles River dummy variable')
pyplot.show()
```

```python
boston_df['binned'] = pd.cut(
    boston_df['AGE'],
    bins=[-np.inf, 35, 70, np.inf],
    labels=["35 and under", "35-70", "Older than 70"],
)
sns.boxplot(x= 'binned' , y='MEDV', data=boston_df)
pyplot.xlabel('Age')
pyplot.title('MEDV variable vs the AGE variable')
pyplot.show()
```

```python
sns.scatterplot(data=boston_df, x = 'NOX' , y= 'INDUS')
pyplot.xlabel('nitric oxides concentration')
pyplot.ylabel('proportion of non-retail business acres per town')
pyplot.title('proportion of non-retail business acres per town vs nitric oxides concentration ')
pyplot.show()
```

```python
pyplot.hist(boston_df['PTRATIO'])
pyplot.xlabel('pupil-teacher ratio by town')
pyplot.title('Histogram for pupil-teacher ratio by town')
pyplot.show() 
```
